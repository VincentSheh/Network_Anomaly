```python
from matplotlib import pyplot as plt
```

```python
duration_list = [
  [1000, 0.00398625],
  [10000, 0.01851],
  [100000, 0.098557875],
  [250000, 0.243153666],
  [500000, 0.47130075],
  [750000, 0.861329292],
  [1000000, 1.029613625],
  [1500000, 2.015812083],
  [2181544, 3.044901458],
]
data_points, durations = zip(*duration_list)

# Creating the plot
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size
plt.plot(data_points, durations, marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Duration vs. Number of Data Points')
plt.xlabel('Number of Data Points')
plt.ylabel('Duration (seconds)')

# Optional: Adding grid for better readability
plt.grid(True)

# Displaying the plot
plt.show()
```
